# Define the Classifiers, ICA

```python
from sklearn.decomposition import FastICA
from sklearn import decomposition

IcaRandom =FastICA(n_components=33)# with RandomForest
X_icaRandom = IcaRandom.fit_transform(X)

IcaXGBoost =FastICA(n_components=31)# with XGBoost
X_icaXGBoost = IcaXGBoost.fit_transform(X)

IcaAdaboost =FastICA(n_components=31)# with AdaBoost
X_icaAdaboost = IcaAdaboost.fit_transform(X)
```

```python
def ICA_Random():
 
   
  modelRandomICA= RandomForestClassifier(bootstrap=True,  max_depth= 60, max_features='sqrt',min_samples_leaf=1,
                          min_samples_split= 5, n_estimators=400)#Random

  scoreICArandom = cross_val_score(modelRandomICA, X_icaRandom, y , cv=cv, n_jobs=-1)

  print('Accuracy of ICA Random: %.3f (%.3f)' % (mean(scoreICArandom), std(scoreICArandom)))


```

```python
def ICA_XGboost():
    

  modelXGBoostICA= XGBClassifier(bootstrap=True,  max_depth= 20, max_features='auto',min_samples_leaf=1,
                         min_samples_split= 2, n_estimators=800)#XGBoost

  scoreICAxgboost = cross_val_score(modelXGBoostICA, X_icaXGBoost, y , cv=cv, n_jobs=-1)

  print('Accuracy of PCA XGboost: %.3f (%.3f)' % (mean(scoreICAxgboost), std(scoreICAxgboost)))
```

```python
def ICA_Adaboost():

  modelAdaBoostICA= AdaBoostClassifier(algorithm='SAMME', learning_rate= 0.3, n_estimators=1600)


  scoreICAadaboost = cross_val_score(modelAdaBoostICA, X_icaAdaboost, y , cv=cv, n_jobs=-1)

  print('Accuracy of PCA AdaBoost: %.3f (%.3f)' % (mean(scoreICAadaboost), std(scoreICAadaboost)))
```
